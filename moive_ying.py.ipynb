```python
import requests
import lxml.html   #导进这个可以运用xpath语法
import csv
from lxml import etree
baseurl="https://movie.douban.com/subject/4864908/comments?start={}&limit=20&sort=new_score&status=P"#这边需要总结出网页的规律
#定义一个函数获取一个网站的数据
def get_one_page(url):
    response=requests.get(url)
    response.status_code
    response.encoding=response.apparent_encoding
    return response.text#也可以用response.content

#定义一个函数解析网站的数据,获取我们所需要的信息
def get_infor(source):
    selector=lxml.html.document_fromstring(source)
    #selector=etree.HTML(source)#这个相当于soup=BeautifulSoup(response.text,"lxml")将网站数据变成一个对象#使用这个函数成功构建一个xpath解析对象
    #获取每一给评论者的评论信息以及特么的名称等
    total_infor=selector.xpath("//div[@class='comment-item']")  #
    #定义一个空列表，用来展示信息的
    movie_list=[]
    #所有的信息都是在total_infor中，我们需要进行展开获得我们所想要的信息
    for each_infor in total_infor:
        autor=each_infor.xpath("div[@class='avatar']/a/@title")
        comment_vote=each_infor.xpath("div[@class='comment']/h3/span[@class='comment_vote']//span[@class='votes']/text()")
        movie_comment=each_infor.xpath("div[@class='comment']/p/span[@class='short']/text()")
        
        #创建一个空的字典来保存我们所想要的电影评论的信息
        movie_dict={}
        movie_dict["autor"]=autor
        movie_dict["comment_vote"]=comment_vote
        movie_dict["movie_comment"]=movie_comment
        #将生成的一个一个的字典形式的信息存储追加导列表中
        movie_list.append(movie_dict)    #此时列表就会变成[dict1{},dict2{}....]
    return movie_list


#定义一个函数用来进行数据的写进scv中
def writedata():
    with open("./yingping.csv","w",encoding="utf-8")as f:
        writer=csv.DictWriter(f,fieldnames=["autor","comment_vote","movie_comment"])
        writer.writeheader()#写进表头
        #传进的参数是一个列表，每个数据是他的一行
        for i in movie_list:
            writer.writerow(i)           
#启动程序
if __name__=="__main__":
    movie_list=[]
    #一个有200条评论，没一页有20条信息，一个有10页，因此只需要循环10次就可以了
    for c in range(11):
        pagelink=baseurl.format(c*20)
        print(pagelink)
        #有了url我们就要获取信息，调用网页获取的信息的函数
        source=get_one_page(pagelink)
        #有了网页信息就可以获得数据，调用获得数据的函数
        movie_list+=get_infor(source)
    print(movie_list)
    writedata()
        
        
    
    
#成功运行。成功写进    
```
